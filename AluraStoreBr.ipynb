### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
Faturamento_por_loja = loja_consolidada.groupby('Loja')['Preço'].sum()
df_Fat_loja = Faturamento_por_loja.to_frame()
#Deixando a coluna preço com o visual de dinheiro
df_Fat_loja['Preço'] = df_Fat_loja.map('R${:,.2f}'.format)
df_Fat_loja = df_Fat_loja.reset_index()
df_Fat_loja
```

```python
# Convertendo a coluna Preço para numérico
df_Fat_loja['Preço'] = pd.to_numeric(df_Fat_loja['Preço'].str.replace(r'[R$,]', '', regex=True), errors='coerce')

# Grafico de Barras com os Faturamentos das lojas
plt.figure(figsize=(10, 6))
plt.bar(df_Fat_loja['Loja'], df_Fat_loja['Preço'])
plt.xlabel('Loja')
plt.ylabel('Preço')
plt.title('Preço por Loja')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Grafico de Pizza para visualizar qual loja contribui mais para o faturamento
plt.figure(figsize=(8, 8))
plt.pie(df_Fat_loja['Preço'], labels=df_Fat_loja['Loja'], autopct='%1.1f%%', startangle=90)
plt.title('Faturamento por Loja')
plt.axis('equal')
plt.show()
```

O gráfico anterior mostra que a loja 4 tem menor participaçao na composiçao do faturamento geral

```python
#Faturamento por loja por ano
loja_consolidada['Data da Compra'] = pd.to_datetime(loja_consolidada['Data da Compra'])
faturamento_loja_ano = loja_consolidada.groupby(['Loja', loja_consolidada['Data da Compra'].dt.year])['Preço'].sum().reset_index()
faturamento_loja_ano
```

```python
# Criando o gráfico de linhas com o faturamento das lojas ao longo dos anos
plt.figure(figsize=(10, 4))

for loja in faturamento_loja_ano['Loja'].unique():
    loja_data = faturamento_loja_ano[faturamento_loja_ano['Loja'] == loja]
    plt.plot(loja_data['Data da Compra'], loja_data['Preço'], label=loja)

plt.xlabel('Ano')
plt.ylabel('Faturamento')
plt.title('Faturamento por Loja ao Longo dos Anos')

# Define os anos desejados para o eixo x
anos_desejados = [2020, 2021, 2022, 2023]
plt.xticks(anos_desejados)

plt.legend()
plt.tight_layout()
plt.show()

```

Pode ser verificado que a loja 1 se mantem como a  loja com maior faturamento na maior parte do tempo, apesar da queda nos 2 últimos anos.

A loja 4 fica como a pior em faturamento desde 2021, em constante queda.

As lojas 2 e 3, tiveram aumento de faturamento desde o inicio do periodo analisado. Ultrapassaram a loja 1 em 2022. E no ultimo ano, onde todas as lojas tiveram queda, acompanharam o comportamento geral, porém mesmo com a queda elas tem um faturamento maior que a loja 4.  

# 2. Vendas por Categoria


```python
vendas_categoria = loja_consolidada.groupby(['Categoria do Produto']).agg(
    Total_vendas=('Preço', 'sum'),
    Quantidade_vendas=('Preço', 'count'),
    Media_preco=('Preço', 'mean'))
vendas_categoria = vendas_categoria.reset_index()
vendas_categoria.sort_values('Total_vendas', ascending=False)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(vendas_categoria['Total_vendas'], labels=vendas_categoria['Categoria do Produto'], autopct='%1.1f%%', startangle=90)
plt.title('Faturamento por Categoria de Produto')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(title="Lojas", loc="best", fontsize=6)
plt.show()
```

As categoras que mais contribuem para o faturamento, são Eletronicos , Eletrodomesticos e Móveis





```python
vendas_categoria_loja = loja_consolidada.groupby(['Categoria do Produto','Loja']).agg(
    Total_vendas=('Preço', 'sum'),
    Quantidade_vendas=('Preço', 'count'),
    Media_preco=('Preço', 'mean'))
vendas_categoria_loja = vendas_categoria_loja.reset_index()
vendas_categoria_loja.sort_values(['Total_vendas','Quantidade_vendas'], ascending=False)
```

```python

import matplotlib.pyplot as plt

for category in vendas_categoria['Categoria do Produto'].unique():
    category_data = vendas_categoria[vendas_categoria['Categoria do Produto'] == category]
    category_data = category_data.sort_values('Total_vendas', ascending=False)

    plt.figure(figsize=(10, 6))
    plt.bar(category_data['Loja'], category_data['Total_vendas'])
    plt.xlabel('Loja')
    plt.ylabel('Total de Vendas')
    plt.title(f'Total de Vendas por Loja - Categoria: {category}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Ranking das lojas que mais vendem eletrodomésticos, eletrônicos e móveis

# Filtrar as categorias desejadas
categorias_interesse = ['eletrodomesticos', 'eletronicos' ,'moveis' ]
vendas_categorias_interesse = vendas_categoria_loja[vendas_categoria_loja['Categoria do Produto'].isin(categorias_interesse)]

# Agrupar por loja e somar as vendas
ranking_lojas = vendas_categorias_interesse.groupby('Loja')['Total_vendas'].sum().sort_values(ascending=False)


print("Ranking das lojas que mais vendem eletrodomesticos, eletronicos e moveis:")
print(ranking_lojas)


plt.figure(figsize=(10, 4))
plt.bar(ranking_lojas.index, ranking_lojas.values)
plt.xlabel('Loja')
plt.ylabel('Total de Vendas')
plt.title('Ranking de Vendas por Loja (eletrodomesticos, eletronicos e moveis)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

As lojas 1,2 e 3 sao as que mais vendem os produtos de maior faturamento. A loja 4 é pioneira de vendas nas categorias de brinqueddos e utilizades domésticas.Que sao as categorias que geram menor faturamento.

```python

#Grafico para entender se teve mudança de comportamento de vendas por categoria ao longo dos anos
sales_by_category_year = loja_consolidada.groupby([loja_consolidada['Data da Compra'].dt.year, 'Categoria do Produto'])['Preço'].sum().reset_index()

plt.figure(figsize=(12, 6))

for category in sales_by_category_year['Categoria do Produto'].unique():
    category_data = sales_by_category_year[sales_by_category_year['Categoria do Produto'] == category]
    plt.plot(category_data['Data da Compra'], category_data['Preço'], label=category)

plt.xlabel('Ano')
plt.ylabel('Total de Vendas')
plt.title('Total de Vendas por Categoria ao Longo dos Anos')
plt.legend(title='Categoria do Produto')
plt.xticks(sales_by_category_year['Data da Compra'].unique())
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
avaliacao_loja = loja_consolidada.groupby('Loja')['Avaliação da compra'].mean()
avaliacao_loja = avaliacao_loja.reset_index()
avaliacao_loja.sort_values('Avaliação da compra', ascending=False)
```

A loja 3 tem a maior média de avaliaçao de compras no geral. Porém como um todo, as lojas não tem muita diferença na avaliaçao dos consumidores.

```python
# Group data by year, store, and calculate the mean of 'Avaliação da Compra'
avaliacao_ano_loja = loja_consolidada.groupby([loja_consolidada['Data da Compra'].dt.year, 'Loja'])['Avaliação da compra'].mean().reset_index()


# Plotting
plt.figure(figsize=(12, 6))
for loja in avaliacao_ano_loja['Loja'].unique():
    loja_data = avaliacao_ano_loja[avaliacao_ano_loja['Loja'] == loja]
    plt.plot(loja_data['Data da Compra'], loja_data['Avaliação da compra'], label=loja)

plt.xlabel('Ano')
plt.ylabel('Avaliação Média da Compra')
plt.title('Avaliação Média da Compra por Ano e Loja')
plt.legend(title='Loja')
plt.grid(True)
plt.xticks(avaliacao_ano_loja['Data da Compra'].unique())
plt.show()
```

```python

avaliacao_categoria_loja = loja_consolidada.groupby(['Loja', 'Categoria do Produto'])['Avaliação da compra'].mean().reset_index()


plt.figure(figsize=(16, 8))

for loja in avaliacao_categoria_loja['Loja'].unique():
    loja_data = avaliacao_categoria_loja[avaliacao_categoria_loja['Loja'] == loja]
    plt.plot(loja_data['Categoria do Produto'], loja_data['Avaliação da compra'], label=loja, marker='o')

plt.xlabel('Categoria do Produto')
plt.ylabel('Avaliação Média')
plt.title('Avaliação Média por Categoria de Produto por Loja')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Loja')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
vendas_produto = loja_consolidada.groupby(['Produto']).agg(
    Total_vendas=('Preço', 'sum'),
    Quantidade_vendas=('Preço', 'count'),
    Media_preco=('Preço', 'mean'))
vendas_produto = vendas_produto.reset_index()
vendas_produto.sort_values('Total_vendas', ascending=False)
```

```python
#Pareto com os proutos que compoem o faturamento
vendas_produto = loja_consolidada.groupby('Produto')['Preço'].sum().sort_values(ascending=False).reset_index()
vendas_produto['Percentual'] = (vendas_produto['Preço'] / vendas_produto['Preço'].sum()) * 100
vendas_produto['Percentual Acumulada'] = vendas_produto['Percentual'].cumsum()

plt.figure(figsize=(12, 6))
plt.bar(vendas_produto['Produto'], vendas_produto['Percentual'], color='skyblue')
plt.plot(vendas_produto['Produto'], vendas_produto['Percentual Acumulada'], color='red', marker='o', linestyle='--')

plt.xlabel('Produto')
plt.ylabel('Percentual de Vendas')
plt.title('Gráfico de Pareto - Percentual de Vendas por Produto')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
vendas_produto['Percentual Acumulada'] = vendas_produto['Percentual'].cumsum()

produtos_80_porcento = vendas_produto[vendas_produto['Percentual Acumulada'] <= 80]

lista_produtos = produtos_80_porcento['Produto'].tolist()

lista_produtos
```

# 5. Frete Médio por Loja


```python
frete_medio_loja = loja_consolidada.groupby('Loja')['Frete'].mean()
frete_medio_loja = frete_medio_loja.reset_index()
frete_medio_loja.sort_values('Frete', ascending=False)
```

```python
frete_loja = loja_consolidada.groupby(['Loja']).agg(
    Total_vendas=('Frete', 'sum'),
    Media_frete=('Frete', 'mean'))
frete_loja = frete_loja.reset_index()
frete_loja.sort_values('Total_vendas', ascending=False)
```

```python
local_compra = loja_consolidada.groupby(['Local da compra']).agg(
    Total_vendas=('Frete', 'sum'),
    Quantidade_vendas=('Local da compra', 'count'),
    Media_frete=('Frete', 'mean'))
local_compra = local_compra.reset_index()
local_compra.sort_values('Total_vendas', ascending=False)
```

```python

local_compra = local_compra.sort_values('Media_frete', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(local_compra['Local da compra'], local_compra['Media_frete'] , color = 'green')
plt.xlabel('Local da Compra')
plt.ylabel('Frete Médio')
plt.title('Frete Médio por Local da Compra')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
local_compra_loja = loja_consolidada.groupby(['Local da compra','Loja']).agg(
    Total_vendas=('Frete', 'sum'),
    Quantidade_vendas=('Local da compra', 'count'),
    Media_frete=('Frete', 'mean'))
local_compra_loja = local_compra_loja.reset_index()
local_compra_loja.sort_values('Total_vendas', ascending=False)
```

```python
# Número de locais de compra distintos atendidos por loja
locais_por_loja = loja_consolidada.groupby('Loja')['Local da compra'].nunique().reset_index()
locais_por_loja.rename(columns={'Local da compra': 'Quantidade de Locais Distintos'}, inplace=True)
locais_por_loja
```

```python
# Top 5 locais de compra com maior venda
top_locais = local_compra.sort_values(by='Total_vendas', ascending=False).head(5)

plt.figure(figsize=(6, 4))
plt.bar(top_locais['Local da compra'], top_locais['Total_vendas'])
plt.xlabel('Local da Compra')
plt.ylabel('Total de Vendas')
plt.title('Top 5 Locais de Compra com Maior Venda')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Lojas que mais vendem por local de compra (considerando apenas os top 5 locais)
for local in top_locais['Local da compra']:
    local_data = local_compra_loja[local_compra_loja['Local da compra'] == local]
    local_data = local_data.sort_values(by='Total_vendas', ascending=False)

    plt.figure(figsize=(6, 4))
    plt.bar(local_data['Loja'], local_data['Total_vendas'])
    plt.xlabel('Loja')
    plt.ylabel('Total de Vendas')
    plt.title(f'Vendas por Loja no Local: {local}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
